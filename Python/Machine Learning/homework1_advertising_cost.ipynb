```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X = df['Total spent']
X
```

```python
X = df['Total spent'].apply(lambda x:[x]).tolist()
X
```

```python
X = np.array(X)
X.shape
```

```python
y = df['sales']
y
```

```python
b = 4.243
w = np.array([0.0487])
```

```python
def compute_cost(X, y, w, b):
    m, n = X.shape
    cost = 0.0
    for i in range(m):
        f_wb_i = np.dot(w, X[i]) + b
        Res = f_wb_i - y[i]
        print(f'Prediction: {f_wb_i}, Residual: {Res}, Residual^2: {(Res)**2}')
        cost = cost + (Res)**2
        
    cost = cost/(2*m)
    return cost
```

```python
cost = compute_cost(X, y, w, b)
print(f'Cost: {cost}')
```
